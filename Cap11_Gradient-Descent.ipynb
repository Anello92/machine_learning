## Gradient Descent

Desafio: 

Encontre os mínimos locais da função y = (x + 5)² começando do ponto x = 3.

```python
from IPython.display import Image
Image("images/grafico.png")
```

Solução: 

Sabemos a resposta apenas olhando para o gráfico. y = (x + 5)² atinge seu valor mínimo quando x = -5 (ou seja, quando x = -5, y = 0). Portanto, x = -5 é o mínimo local e global da função.

### Aplicando Gradient Descent

```python
# O algoritmo inicia com o parâmetro x=3
cur_x = 3 

# Learning rate
rate = 0.01 

# Variável "precision" define quando parar o algoritmo
# Parar o loop quando a diferença entre os valores de x em 2 iterações consecutivas for menor que 0,000001 
# ou quando o número de iterações exceder 10 mil de "max_iters"
precision = 0.000001 

# Inicializa o contador do passo anterior
previous_step_size = 1 

# Número máximo de iterações
max_iters = 10000 

# Contador de iterações já que vamos até max_iters de 10.000
iters = 0 

# Gradiente da função anônima
# Valor de x para 2 que multiplica x+5 em df 
df = lambda x: 2*(x+5) 
```

```python
# TREINAMENTO 

# Se por previous for maior que precision & n° iteração menor que o valor máximo de iterações, continuamos o loop
while previous_step_size > precision and iters < max_iters:
# Enquanto essa condição for verdadeira continuo excutando o loop

    # Armazena o valor corrente de x em prev_x
    prev_x = cur_x 
    
    # Aplica o Gradient descent 
    # Valor corrente de x - taxa de aprendizado * gradient do valor de x
    cur_x = cur_x - rate * df(prev_x) 
    
    # Altera o valor de x
    # Valor corrent - Valor prévio assim como cálculo da Taxa de Erros
    previous_step_size = abs(cur_x - prev_x) 
    
    # Incrementa o número de iterações
    iters = iters + 1 
    
    # Imprime as iterações
    print("Iteration", iters,"\nValor de x igual a ", cur_x) 


print("\nO mínimo local da função ocorre em: ", cur_x)
```
