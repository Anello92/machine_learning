# MSE - Erro Quadrado Médio

Similar ao MAE - Erro Absoluto Médio. 

É talvez a métrica mais simples e comum para a avaliação de regressão, mas também provavelmente a menos útil. O MSE basicamente mede o erro quadrado médio de nossas previsões. Para cada ponto, calcula a diferença quadrada entre as previsões e o valor real da variável alvo e, em seguida, calcula a média desses valores.

Quanto maior esse valor, pior é o modelo. Esse valor nunca será negativo, já que estamos elevando ao quadrado os erros individuais de previsão, mas seria zero para um modelo perfeito.

```python
# MSE - Mean Squared Error (É uma taxa de ERRO)
# Similar ao MAE, fornece a magnitude do erro do modelo.
# Quanto maior, pior é o modelo!
# Ao extrairmos a raiz quadrada do MSE convertemos as unidades de volta ao original, 
# o que pode ser útil para descrição e apresentação. Isso é chamado RMSE (Root Mean Squared Error)

# Import dos módulos
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error # Função MSE
from sklearn.linear_model import LinearRegression

# Carregando os dados
arquivo = 'data/boston-houses.csv'
colunas = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO','B', 'LSTAT', 'MEDV']
dados = read_csv(arquivo, delim_whitespace = True, names = colunas)
array = dados.values

# Separando o array em componentes de input e output
X = array[:,0:13] # Preditoras todas as colunas exceto a última
Y = array[:,13] # Target todas as linhas última coluna 

# Divide os dados em treino e teste com a função train_test_split()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 5)

# Criando o modelo
modelo = LinearRegression() # Chamando o modelo de Regressão Linear

# Treinando o modelo
modelo.fit(X_train, Y_train) # Treinamento do modelo de Regressão 

# Fazendo previsões com conjunto de teste
Y_pred = modelo.predict(X_test)

# Resultado      Y_test: Real Histórico   Y_pred:Valores calculados 
mse = mean_squared_error(Y_test, Y_pred)
print("O MSE do modelo é:", mse)

# O ideal é que depois do MSE do Modelo de 28, fosse feito um trabalho em cima desse conjunto de dados
# Pré-Processamento, Transformação de Variáveis, Normalização, Padronização
# Tentando diminuir o MSE que é uma taxa de ERRO
```

# MAE - Erro Absoluto Médio

```python
# MAE - Mean Absolute Error
# É a soma da diferença absoluta entre previsões e valores reais.

# Fornece uma ideia de quão erradas estão nossas previsões.

# Valor igual a 0 indica que não há erro, sendo RARA a previsão perfeita.

# Import dos módulos
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error # Esta função que muda
from sklearn.linear_model import LinearRegression

# Carregando os dados
arquivo = 'data/boston-houses.csv'
colunas = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO','B', 'LSTAT', 'MEDV']
dados = read_csv(arquivo, delim_whitespace = True, names = colunas)
array = dados.values

# Separando o array em componentes de input e output
X = array[:,0:13]
Y = array[:,13]

# Divide os dados em treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 5)

# Criando o modelo
modelo = LinearRegression()

# Treinando o modelo
modelo.fit(X_train, Y_train)

# Fazendo previsões
Y_pred = modelo.predict(X_test)

# Resultado, depois que calculado as previsões, passamos como parâmetro
  # Y_test - valor real histórico  Y_Pred - valor preditivos
mae = mean_absolute_error(Y_test, Y_pred)
print("O MAE do modelo é:", mae)

# DETALHE IMPORTANTE, não podemos comparar MAE com MSE
# Temos que comaparar 2 modelos com a MESMA MÉTRICA
```

# Algoritmos de Regressão

## Regressão Linear

Regressão Linear é o algoritmo mais simples de toddos.

Temos duas variantes principais: 
- Regressão Linear Simples: uma variável de entrada
- Regressão Linear Múltipla: muitas variáveis de entrada

Assume que os dados estão em Distribuição Normal e também assume que as variáveis são relevantes para a construção do modelo e que não sejam colineares, ou seja, variáveis com alta correlação (cabe a você, Cientista de Dados, entregar ao algoritmo as variáveis realmente relevantes).

```python
# Import dos módulos
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression #Regressão linear

# Carregando os dados 
arquivo = 'data/boston-houses.csv'
colunas = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO','B', 'LSTAT', 'MEDV']
dados = read_csv(arquivo, delim_whitespace = True, names = colunas)
array = dados.values

# Separando o array em componentes de input e output
X = array[:,0:13]
Y = array[:,13] #Target preço da casa

# Divide os dados em treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 5)

# Criando o modelo em uma linha com Scikitlearn
modelo = LinearRegression()

# Treinando o modelo
modelo.fit(X_train, Y_train)

# Fazendo previsões
Y_pred = modelo.predict(X_test)

# Resultado
mse = mean_squared_error(Y_test, Y_pred)
print("O MSE do modelo é:", mse)

# É possível reduzir essa taxa de erro apenas mudando o algoritmo?
```

## Ridge Regression

Quando construimos o modelo, precisamos automatizar o processo. Para automatizar o processo precisamos usar o gradiente descendente, por exemplo, um algoritmo para treinar o modelo.


Extensão para a regressão linear onde a loss function é modificada para minimizar a complexidade do modelo.

- Função de Perda
- Função de Erro
- Função de Custo

Para que se possa usar o gradiente descendente, é necessário escolher uma Função de Custo. O gradiente descendente que é um dos muitos algoritmos para treinamento, ele OTIMIZA essa Função de Custo, reduzindo essa função tanto quanto possível. 

Se reduzirmos a Função de Custo, diminuímos a Taxa de Erro do modelo. Em linhas gerais Machine Learning trata-se de um problema de otimização, OTIMIZAR a Função Custo. Portanto temos:

- Algoritmo de Regressão Linear
- Algoritmo de Otimização "Gradiente Descendente"
- Função de Custo

Colocamms tudo isso junto e treinamos o modelo. No final seremos capazez de encontrar a MELHOR RELAÇÃO MATEMÁTICA entre os laços de entrada e de saída. 

```python
# Import dos módulos
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge

# Carregando os dados
arquivo = 'data/boston-houses.csv'
colunas = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO','B', 'LSTAT', 'MEDV']
dados = read_csv(arquivo, delim_whitespace = True, names = colunas)
array = dados.values

# Separando o array em componentes de input e output
X = array[:,0:13]
Y = array[:,13]

# Divide os dados em treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 5)

# Criando o modelo
modelo = Ridge()

# Treinando o modelo
modelo.fit(X_train, Y_train)

# Fazendo previsões
Y_pred = modelo.predict(X_test)

# Resultado
mse = mean_squared_error(Y_test, Y_pred)
print("O MSE do modelo é:", mse)
```

## Otimização do Modelo - Ajuste de Parâmetros

Todos os algoritmos de Machine Learning são parametrizados, o que significa que você pode ajustar a performance do seu modelo preditivo, através do tuning (ajuste fino) dos parâmetros. Seu trabalho é encontrar a melhor combinação entre os parâmetros em cada algoritmo de Machine Learning. Esse processo também é chamado de Otimização Hyperparâmetro. O scikit-learn oferece dois métodos para otimização automática dos parâmetros: Grid Search Parameter Tuning e Random Search Parameter Tuning. 

### Grid Search Parameter Tuning

Este método realiza metodicamente combinações entre todos os parâmetros do algoritmo, criando um grid. Vamos experimentar este método utilizando o algoritmo de Regressão Ridge. No exemplo abaixo veremos que o valor 1 para o parâmetro alpha atingiu a melhor performance.

```python
# Import dos módulos
from pandas import read_csv
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

# Carregando os dados
arquivo = 'data/boston-houses.csv'
colunas = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO','B', 'LSTAT', 'MEDV']
dados = read_csv(arquivo, delim_whitespace = True, names = colunas)
array = dados.values

# Separando o array em componentes de input e output
X = array[:,0:8]
Y = array[:,8]

# Definindo os valores que serão testados
valores_alphas = np.array([1,0.1,0.01,0.001,0.0001,0])
valores_grid = dict(alpha = valores_alphas)

# Criando o modelo
modelo = Ridge()

# Criando o grid
grid = GridSearchCV(estimator = modelo, param_grid = valores_grid)
grid.fit(X, Y)

# Print do resultado dos melhores parâmetro para o algoritmo
print("Melhores Parâmetros do Modelo:\n", grid.best_estimator_)
```

### Random Search Parameter Tuning

Este método gera amostras dos parâmetros dos algoritmos a partir de uma distribuição randômica uniforme para um número fixo de interações. Um modelo é construído e testado para cada combinação de parâmetros. Neste exemplo veremos que o valor muito próximo de 1 para o parâmetro alpha é o que vai apresentar os melhores resultados.

```python
# Import dos módulos
from pandas import read_csv
import numpy as np
from scipy.stats import uniform
from sklearn.linear_model import Ridge
from sklearn.model_selection import RandomizedSearchCV

# Carregando os dados
arquivo = 'data/boston-houses.csv'
colunas = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO','B', 'LSTAT', 'MEDV']
dados = read_csv(arquivo, delim_whitespace = True, names = colunas)
array = dados.values

# Separando o array em componentes de input e output
X = array[:,0:8]
Y = array[:,8]

# Definindo os valores que serão testados
valores_grid = {'alpha': uniform()}
seed = 7

# Criando o modelo
modelo = Ridge()
iterations = 100
rsearch = RandomizedSearchCV(estimator = modelo, 
                             param_distributions = valores_grid, 
                             n_iter = iterations, 
                             random_state = seed)
rsearch.fit(X, Y)

# Print do resultado
print("Melhores Parâmetros do Modelo:\n", rsearch.best_estimator_)
```
